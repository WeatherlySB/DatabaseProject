```python
import requests
import pandas as pd
import os
from PIL import Image
from io import BytesIO

def download_and_show_image(url, headers):
    try:
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            img = Image.open(BytesIO(response.content))
            img.show()
            print("Image displayed successfully.")
        else:
            print(f"Failed to download {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"An error occurred while downloading the image: {e}")

def show_local_image(path):
    if os.path.exists(path):
        img = Image.open(path)
        img.show()
        print("Local image displayed successfully.")
    else:
        print(f"Local image not found at path: {path}")

if __name__ == "__main__":
    headers = {
        "User-Agent": "Gecko/20100101",
    }

    # Load the CSV
    df = pd.read_csv("../fish_data/fishbase_img.csv")

    # Get the SpecCode input
    target_spec_code = int(input("Enter the SpecCode: "))

    # Look for the image path associated with the SpecCode
    match = df[(df["SpecCode"] == target_spec_code) & df["PicPreferredName"].notna()]

    if not match.empty:
        img_path = match.iloc[0]["PicPreferredName"]
        if img_path.startswith("https"):
            print(f"Downloading image from: {img_path}")
            download_and_show_image(img_path, headers)
        else:
            print(f"Showing local image from: {img_path}")
            show_local_image(img_path)
    else:
        print(f"No image found for SpecCode {target_spec_code}.")
```

```python

# Create the output directory if it doesn't exist
import pandas as pd


# Define the base URL
base_url = "https://fishbase.org/images/species/"
new_url = "https://www.fishbase.se/images/species/"

# Load your CSV file
df = pd.read_csv("fishbase_img.csv")

# Only modify non-null values in the PicPreferredName column
df['PicPreferredName'] = df['PicPreferredName'].apply(
    lambda x: new_url + x if pd.notna(x) else x
)

#if 'PicPreferredName' in df.columns:
    # Extract just the filename from the URL path
 #   df['PicPreferredName'] = df['PicPreferredName'].apply(
  #      lambda url: url.replace(base_url, "") if isinstance(url, str) else url
   # )

# Save the modified dataframe back to CSV
df.to_csv("fishbase_img.csv", index=False)
```

```python
# When using DuckDB, we need to convert to pandas or use DuckDB's SQL interface
# Method 1: Convert to pandas DataFrame first
image_data = df.select("SpecCode", "PicPreferredName").filter("PicPreferredName IS NOT NULL").to_df()

# Additional pandas filtering for empty strings
image_data = image_data[image_data['PicPreferredName'].apply(
    lambda x: isinstance(x, str) and x.strip() != ''
)]

# Base URL for the images
base_url = "https://fishbase.org/images/species/"

# Counter for tracking progress
total_images = len(image_data)
successful_downloads = 0
failed_downloads = 0

print(f"Found {total_images} images to download.")

# Download each image
for index, row in image_data.iterrows():
    spec_code = row['SpecCode']
    image_name = row['PicPreferredName']
    
    # Create the full URL
    img_url = urljoin(base_url, image_name)
    
    # Destination path
    dest_path = output_dir / image_name
    
    try:
        print(f"Downloading SpecCode {spec_code}, Image: {image_name}")
        
        # Make the request
        response = requests.get(img_url, stream=True)
        response.raise_for_status()  # Raise an exception for 4XX/5XX responses
        
        # Save the image
        with open(dest_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        
        successful_downloads += 1
        print(f"Successfully downloaded {image_name}")
        
    except Exception as e:
        failed_downloads += 1
        print(f"Failed to download {image_name}: {e}")
    
    # Print progress
    if index > 0 and index % 10 == 0:
        print(f"Progress: {index}/{total_images} images processed")

print("\nDownload Summary:")
print(f"Total images: {total_images}")
print(f"Successfully downloaded: {successful_downloads}")
print(f"Failed downloads: {failed_downloads}")
```
